```python
reuter_clean=[]
for i in reuter.sentence_text:
    #print(i)
    i=re.sub("[^a-zA-Z' ]",'',i)
    i=i.lower()
    reuter_clean.append(i)
    #print(i)
```

```python
#print(reuter_clean)
```

```python
all_words=" ".join(reuter_clean).split()
words_dict={}
for word in all_words:
    if word in words_dict:
        words_dict[word]+=1
    else:
        words_dict[word]=1
        
```

```python
word_df=pd.DataFrame({'word':list(words_dict.keys()),'count':list(words_dict.values())})
word_df=word_df.sort_values(by=['count'])
word_df.reset_index(inplace = True, drop=True)
word_df.head()
```

```python
dataset=pd.DataFrame()
dataset['sentence']=reuter_clean
dataset.head()
```

```python
dataset['sentence'][0].split()
```

```python
def create_unigram(sentence):
    token=sentence.split()
    unigram=[]
    for i in range(len(token)):
        unigram.append(token[i:i+1])
    return unigram    
```

```python
def create_bigram(sentence):
    token=sentence.split()
    bigram=[]
    for i in range(len(token)-1):
        bigram.append(token[i:i+2])
    return bigram    
```

```python
def create_trigram(sentence):
    token=sentence.split()
    trigram=[]
    for i in range(len(token)-2):
        trigram.append(token[i:i+3])
    return trigram    
```

```python
final_unigram=[]
for i in range(dataset.shape[0]):
    final_unigram.append(create_unigram(dataset['sentence'][i]))
dataset['unigram']=final_unigram    
```

```python
final_bigram=[]
for i in range(dataset.shape[0]):
    final_bigram.append(create_bigram(dataset['sentence'][i]))
dataset['bigram']=final_bigram    
```

```python
final_trigram=[]
for i in range(dataset.shape[0]):
    final_trigram.append(create_trigram(dataset['sentence'][i]))
dataset['trigram']=final_trigram 
```

```python
from collections import Counter, defaultdict
model=defaultdict(lambda: defaultdict(lambda:0))
for i in range(dataset.shape[0]):
    for w1,w2,w3 in create_trigram(dataset['sentence'][i]):
        model[(w1,w2)][w3]+=1
```

```python
unigram_dict = {}
for i in tqdm(range(dataset.shape[0])):
    # add word-count pair to the dictionary
    for word in dataset['unigram'][i]:   
        # check if the word is already in dictionary 
        if word[0] in unigram_dict:
            # increment count of word by 1 
            unigram_dict[word[0]] = unigram_dict[word[0]] + 1
        else:
            # add the word to dictionary with count 1 
            unigram_dict[word[0]] = 1
```

```python
unigram_dict
```

```python
counts = Counter(unigram_dict)
counts
```

```python
total_count = len(unigram_dict)
total_count
```

```python
for word in counts:
    counts[word] /= float(total_count)

counts
```

```python
for w1_w2 in model:
    total_count = float(sum(model[w1_w2].values()))
    for w3 in model[w1_w2]:
        model[w1_w2][w3] /= total_count
```
