KNN

```python


train_Sx,test_Sx,train_Sy,test_Sy=train_test_split(Sx,Sy,random_state=56,test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as MSE
```

```python
regr=KNN(n_neighbors=3)
regr.fit(train_Sx,train_Sy)
```

```python
test_predict=regr.predict(test_Sx)
k=MSE(test_predict,test_Sy)
```

```python
print('test MSE',k)
```

Elbow for classifier

```python
def Elbow(k):
    test_MSE=[]
    for i in k:
        regr=KNN(n_neighbors=i)
        regr.fit(train_Sx,train_Sy)
        tmp=regr.predict(test_Sx)
        tmp=MSE(tmp,test_Sy)
        test_MSE.append(tmp)
        
    return test_MSE
```

```python
k=range(1,40)
```

```python
test=Elbow(k)
```

```python
plt.plot(k,test)
plt.xlabel('KNeighbors')
plt.ylabel('Test MSE')
plt.title('Elbow curve for test')
```

```python
regr=KNN(n_neighbors=4)
regr.fit(train_Sx,train_Sy)
test_predict=regr.predict(test_Sx)
k=MSE(test_predict,test_Sy)
print('Test MSE',k)
```

```python
regr.score(train_Sx,train_Sy)*100
```

```python
regr.score(test_Sx,test_Sy)*100
```

```python
names=["LinearRegressor","DecisionTree","SVM","KNN"]

classifiers=[
    LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False),
    DecisionTreeRegressor(max_depth=None),
    SVR(kernel='rbf',degree=3,C=1.0),
    KNN(4)]
```

```python
scores=[]
for name,clf in zip(names,classifiers):
    clf.fit(train_Sx,train_Sy)
    score=clf.score(test_Sx,test_Sy)
    scores.append(score)
```

```python
scores
```

```python
df=pd.DataFrame()
df['name']=names
df['accuracy score']=scores
df
```
